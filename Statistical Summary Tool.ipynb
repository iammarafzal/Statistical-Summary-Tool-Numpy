```python
import numpy as np
```

##### 1-dimensional Array

```python
user_input = input("Enter numbers separting by commas: ")

try:
    arr_1d = np.array([float(x) for x in user_input.split(',')])
except ValueError:
    print("Invalid input. Please enter numbers only.")
```

```python
print("Count: ", np.count_nonzero(arr_1d))
```

```python
print("Mean (Average): ", np.mean(arr_1d))
```

```python
print("Median (middle value): ", np.median(arr_1d))
```

```python
print("Min Value: ", arr_1d.min())
```

```python
print("Max Value: ", np.max(arr_1d))
```

```python
print("Range: ", np.max(arr_1d) - np.min(arr_1d))
```

```python
print("Variance: ", np.var(arr_1d))
```

```python
print("Standard Deviation: ", np.std(arr_1d))
```

```python
print("Sum: ", np.sum(arr_1d))
```

```python
def statistical_summary(arr):
    print("Array:", arr)
    print("Count of Non-Zero Elements:", np.count_nonzero(arr))
    print("Mean:", np.mean(arr))
    print("Median:", np.median(arr))
    print("Standard Deviation:", np.std(arr))
    print("Variance:", np.var(arr))
    print("Minimum:", np.min(arr))
    print("Maximum:", np.max(arr))
    print("25th Percentile:", np.percentile(arr, 25))
    print("75th Percentile:", np.percentile(arr, 75))

print("----- Statistical Summary -----")
statistical_summary(arr_1d)
```

```python
from matplotlib.pylab import plt
def generate_visualizations(arr):
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))

    # Histogram
    axs[0, 0].hist(arr, bins=5, edgecolor='black', color='skyblue')
    axs[0, 0].set_title("Histogram")
    axs[0, 0].set_xlabel("Values")
    axs[0, 0].set_ylabel("Frequency")

    # Boxplot
    axs[0, 1].boxplot(arr, vert=False)
    axs[0, 1].set_title("Boxplot")
    axs[0, 1].set_xlabel("Values")

    # Pie Chart (only if few unique values)
    unique_vals, counts = np.unique(arr, return_counts=True)
    if len(unique_vals) <= 8:
        axs[1, 0].pie(counts, labels=unique_vals, autopct='%1.1f%%', startangle=90)
        axs[1, 0].set_title("Pie Chart of Values")
    else:
        axs[1, 0].text(0.5, 0.5, 'Too many unique values for pie chart', ha='center', va='center')
        axs[1, 0].set_title("Pie Chart Skipped")

    # Line Plot
    axs[1, 1].plot(arr, marker='o', linestyle='-', color='orange')
    axs[1, 1].set_title("Line Plot (Index vs Value)")
    axs[1, 1].set_xlabel("Index")
    axs[1, 1].set_ylabel("Value")

    plt.tight_layout()
    plt.show()
    plt.savefig("statistical_summary_plots.png", dpi=300)


generate_visualizations(arr_1d)
```

##### 2-dimensional Array

```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n ", arr_2d)
```

```python
print("Count: ", np.count_nonzero(arr_2d))
```

```python
print("Mean: ", np.mean(arr_2d))
print("Mean Column: ", np.mean(arr_2d, axis=0))
print("Mean Row: ", np.mean(arr_2d, axis=1))
```

```python
print("Median: ", np.median(arr_2d))
print("Median Column: ", np.median(arr_2d, axis=0))
print("Median Row: ", np.median(arr_2d, axis=1))
```

```python
print("Min: ", np.min(arr_2d))
print("Min Column: ", np.min(arr_2d, axis=0))
print("Min Row: ", np.min(arr_2d, axis=1))
```

```python
print("Range: ", np.max(arr_2d) - np.min(arr_2d))
```

```python
print("Variance: ", np.var(arr_2d))
```

```python
print("Standard Deviation: ", np.std(arr_2d))
```

```python
print("Sum: ", np.sum(arr_2d))
print("Sum Column: ", np.sum(arr_2d, axis=0))
print("Sum Row: ", np.sum(arr_2d, axis=1))
```
